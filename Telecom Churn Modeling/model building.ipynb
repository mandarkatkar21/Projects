```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('cleaned_file.xlsx')
df.tail()
```

```python
df[df['churn']=='yes']
```

```python
df.info()
```

```python
# Independent and dependent variables
X=df.iloc[:,:-1]
y=df['churn']
```

```python
numerical_clm_after_cleaning=[feature for feature in X.columns if X[feature].dtype!='object']
print(numerical_clm_after_cleaning)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
transformer=ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(sparse_output=False,drop='first'),['voice.plan','intl.plan']),
    ('tnf2',StandardScaler(),numerical_clm_after_cleaning)
                            ],remainder='passthrough')
```

```python
X_train
```

```python
X_train1=transformer.fit_transform(X_train)
```

```python
X_train1
```

```python
X_test1=transformer.transform(X_test)
```

```python
label=LabelEncoder()
```

```python
y_train1=label.fit_transform(y_train)
y_train1
```

```python
y_test1=label.transform(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix,precision_score, recall_score, f1_score, roc_auc_score,roc_curve 
```

```python
models={
    "Logisitic Regression":LogisticRegression(),
    "Support Vector Machine":SVC(),
    "Random Forest":RandomForestClassifier(),
    "XG Boost":XGBClassifier()
}
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train1, y_train1) # Train model

    # Make predictions
    y_train_pred = model.predict(X_train1)
    y_test_pred = model.predict(X_test1)

    # Training set performance
    model_train_accuracy = accuracy_score(y_train1, y_train_pred) # Calculate Accuracy
    model_train_f1 = f1_score(y_train1, y_train_pred, average='weighted') # Calculate F1-score
    model_train_precision = precision_score(y_train1, y_train_pred) # Calculate Precision
    model_train_recall = recall_score(y_train1, y_train_pred) # Calculate Recall
    model_train_rocauc_score = roc_auc_score(y_train1, y_train_pred)


    # Test set performance
    model_test_accuracy = accuracy_score(y_test1, y_test_pred) # Calculate Accuracy
    model_test_f1 = f1_score(y_test1, y_test_pred, average='weighted') # Calculate F1-score
    model_test_precision = precision_score(y_test1, y_test_pred) # Calculate Precision
    model_test_recall = recall_score(y_test1, y_test_pred) # Calculate Recall
    model_test_rocauc_score = roc_auc_score(y_test1, y_test_pred) #Calculate Roc


    print(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Accuracy: {:.4f}".format(model_train_accuracy))
    print('- F1 score: {:.4f}'.format(model_train_f1))
    
    print('- Precision: {:.4f}'.format(model_train_precision))
    print('- Recall: {:.4f}'.format(model_train_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_train_rocauc_score))

    
    
    print('----------------------------------')
    
    print('Model performance for Test set')
    print('- Accuracy: {:.4f}'.format(model_test_accuracy))
    print('- F1 score: {:.4f}'.format(model_test_f1))
    print('- Precision: {:.4f}'.format(model_test_precision))
    print('- Recall: {:.4f}'.format(model_test_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_test_rocauc_score))

    
    print('='*35)
    print('\n')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100,120,150,170],          
    'learning_rate': [ 0.1, 0.2,0.3],      
    'max_depth': [5,6],                  
    'colsample_bytree': [0.9, 1.0]           
}
```

```python
neg_class_count = np.sum(y_train1 == 0)
pos_class_count = np.sum(y_train1 == 1)

# Calculate the scale_pos_weight
scale_pos_weight = neg_class_count / pos_class_count
print(f"Scale Pos Weight: {scale_pos_weight}")
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(
    estimator=XGBClassifier(scale_pos_weight=scale_pos_weight), 
    param_grid=param_grid, 
    cv=5,                
    scoring='accuracy',   
    n_jobs=-1,           
    verbose=2
)

grid_search.fit(X_train1, y_train1)

print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

```python
best_model=XGBClassifier(colsample_bytree= 1.0, learning_rate=0.1, max_depth= 6, n_estimators= 100, reg_alpha= 0.3, reg_lambda= 1,scale_pos_weight=scale_pos_weight)
```

```python
best_model.fit(X_train1,y_train1)
```

```python
y_pred_train=best_model.predict(X_train1)
y_pred_test=best_model.predict(X_test1)
```

```python
print('Classification report for Training data')
print(classification_report(y_train1,y_pred_train))

print('Classification report for Testing data')
print(classification_report(y_test1,y_pred_test))
```

```python
print(confusion_matrix(y_test1,y_pred_test))
```
